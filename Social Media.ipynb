## Data Cleaning

```python
# For numerical columns, we can fill missing values with the median
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
data[numerical_columns] = data[numerical_columns].fillna(data[numerical_columns].median())
```

```python
# For categorical columns, fill missing values with the mode (most frequent value)
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    data[column] = data[column].fillna(data[column].mode()[0])
```

```python
# Standardize 'Location' by correcting known typo
data['Location'] = data['Location'].replace({'Barzil': 'Brazil'})
```

```python
data['Profession'] = data['Profession'].replace({
    'driver': 'Driver', 
    'Labor/Worker': 'Labor Worker'
})
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Convert binary categorical columns to numeric 

binary_columns = ['Owns Property', 'Debt']  
label_encoder = LabelEncoder()
for col in binary_columns:
    data[col] = label_encoder.fit_transform(data[col].astype(str))
```

```python
print("Remaining missing values in each column (should be zero if fully cleaned):")
print(data.isnull().sum())
```

```python
# Displaying the first few rows of the cleaned dataset to verify changes

print(data.head())
```

## Key Insights

## 1. Operations Team: Supply Chain Insights
Insights and Recommendations:
1.User Distribution by Location:

Insight: The highest concentration of users is found in locations such as India, Pakistan, and Indonesia.

Recommendation: Prioritize inventory and logistics support in these regions, ensuring sufficient stock and resources are available to meet demand in these high-user-density areas.

2.Income Distribution by Location:

Insight: Locations like the United States, Germany, and Japan have the highest average income levels, suggesting a greater purchasing power in these regions.

Recommendation: Allocate higher-value inventory to regions with greater purchasing power and consider premium products or services for these areas.

3.Platform Usage by Location:

Insight: Instagram is particularly popular in countries such as Brazil and Pakistan, while YouTube is more popular in the United States and India.

Recommendation: Tailor product offerings based on platform preference in each location. For example, promote visually engaging products on Instagram in Brazil and educational or informational products on YouTube in the United States.

4.Device Type and Connection Type by Location:

Insight: Mobile devices dominate across most locations, with Wi-Fi being more common in urban areas, while mobile data usage is prevalent in rural areas.

Recommendation: Focus on mobile-friendly content and ensure smooth functionality on mobile networks, particularly in rural regions where mobile data usage is higher.

## 2. Sales Team: Vendor Collaboration and Offer Optimization on Flipkart
Insights and Recommendations:

1.User Engagement and Preferences by Platform:

Insight: Instagram and YouTube users show the highest engagement levels.

Recommendation: Focus vendor offers on these platforms to capture the most engaged users. Consider exclusive deals or time-sensitive offers on Flipkart targeted at Instagram and YouTube users.

2.Video Category Preferences:

Insight: Video categories such as Trends, Life Hacks, and Entertainment have higher engagement.

Recommendation: Collaborate with vendors to create offers that align with popular categories. For example, trend-based offers or "how-to" product bundles can resonate well with users interested in Life Hacks.

3. Time Spent and Frequency:

Insight: Users who spend above-average time on the platform and frequently log in are more likely to convert.

Recommendation: Target high-engagement users with exclusive offers, such as personalized discount codes, to increase conversions on Flipkart. Incentives for frequent users could also include loyalty rewards.

4. Income and Purchase Propensity:

Insight: Higher-income users (from locations like Germany and Japan) are more likely to make larger purchases.

Recommendation: Segment offers based on income, promoting premium and high-value products to higher-income users, while offering budget-friendly or value-based products to mid-income segments.

## 3. Marketing Team: Customized Launch Offers
Insights and Recommendations:

1.User Demographics Segmentation:

Insight: Age groups 18-30 and 30-45 form the largest user base, with professionals and students being predominant occupations.

Recommendation: Tailor launch offers to target younger users with trend-focused, fast-moving products, while professionals may respond well to productivity or time-saving products.

2. Platform Engagement:

Insight: Users on Instagram and YouTube have the highest engagement.

Recommendation: Invest in visually appealing, story-driven launch offers on Instagram and YouTube to capture these highly engaged users.

3.Device Type Optimization:

Insight: Smartphones are the most commonly used devices across all regions.

Recommendation: Ensure all launch offers are mobile-optimized, with simple, fast-loading designs. Consider app-based launch campaigns to enhance engagement with smartphone users.

4. Addiction Level and Self-Control Metrics:

Insight: Users with lower self-control and higher addiction levels are more likely to engage with impulsive offers.

Recommendation: Use limited-time discounts, flash sales, and countdowns to drive immediate action from users with higher addiction scores. These offers can create a sense of urgency, increasing conversions.

5. Watch Time Analysis:

Insight: Users are most active during the evening (6:00 PM - 9:00 PM).

Recommendation: Schedule launch offer notifications and promotions to go live in the evening, aligning with peak user activity times to maximize visibility and engagement.
